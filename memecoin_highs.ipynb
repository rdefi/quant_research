### Pattern:
##### The pattern we see in both DOGE rallies (Apr and May 2021) and the second SHIB rally (Oct 2021) is that a top comes when:
##### 1. A candle with a large relative high-low range to previous sessions (drastic increase in realized vol)
##### 2. A candle with a proportionally large upper wick, ie high minus close is large relative to high minus low

##### Once a wick with those characteristics appeared, the wick from the next trading session traded up near those highs
##### an opportunity to sell at the previous day's highs 

### Trader Psychology of this Pattern:
##### Large candles with a small upper wick show that the bid continues into the close after an extreme high is printed in that session.
##### This is bullish and suggests the rally is likely not over, especially in memecoins which have lots of irrationality and fear of missing out.
##### But when a high is reached and then a subsequent close is much lower, it shows that aggressive bid into the high cannot be sustained
##### and thus increases the likelihood of a top being put in, especially if price action has been parabolic leading into that session.

### Strategy:

##### Can do one of two things:
##### 1. Once we see a large candle with a large upper wick, place offers between the wick's close and high to (hopefully) get paid in the next session
#####    (effectiveness of this is subject to change in future rallies of other coins)
##### 2. Once we see a large candle with a large upper wick, keep an eye on the next session -
#####    if the high from the previous day begins to see some rejection, then start selling
